```python
import qutip as qt
import numpy as np

b = qt.Bloch()
b.clear()
vector = [0,0,1]
b.add_vectors(vector)

b.sphere_alpha = 0
b.save('../presentation/images/test_bloch.pdf')
b.show()
```

```python
b.clear()
b.add_vectors([np.sin(0.2), 0, np.cos(0.2)])
# pnts_x = np.sin(0.2)*np.cos(0.2)
# pnts_y = np.sin(0.2)*np.sin(0.2)

# pnts_y = [np.sin(phi) for phi in np.linspace(0, 3*np.pi/2, 10)]
# pnts_x = [np.cos(phi) for phi in np.linspace(0,3*np.pi/2,10)]
phi = np.linspace(0,np.pi,10)
pnts_x = np.cos(phi)*np.sin(0.2)
pnts_y = np.sin(phi)*np.sin(0.2)
pnts_z = [np.cos(0.2)]*10
b.add_points([pnts_x, pnts_y, pnts_z])
b.save('../presentation/images/bloch_z_rot.pdf')
b.show()
```

```python

b.clear()
phi2=np.linspace(0,np.pi/2,5)
pnts_x = [0]*5
pnts_y = -np.sin(phi2)
pnts_z = np.cos(phi2)
for i in range(5):
    b.add_vectors([pnts_x[i], pnts_y[i], pnts_z[i]])

b.sphere_alpha = 0
b.vector_color = ['g']
b.save('../presentation/images/bloch_x_rot.pdf')
b.show()
```
